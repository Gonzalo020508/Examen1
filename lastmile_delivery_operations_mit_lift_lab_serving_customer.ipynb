**DESVIACION ESTANDAR PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#usamos dtypes para saber el tipo de dato
data.dtypes
```

```python
#realizamos ahora un filtro por columnas de las que son de tipo cuantitativas 
datoscuantitativos=data1.iloc[ : , [4,9,11,12,14,15,16,19,20,24]]
datoscuantitativos
```

```python
#corroboramos valores nulos sabiendo que no hay ninguno
valores_nulos=datoscuantitativos.isnull().sum()
valores_nulos
```

```python
#realizamos una copia del dataframe
data3=datoscuantitativos.copy()
```

```python
##hacemos uso de desviacion estandar puesto que eliminarlos no es una solucion y permite ver que tan alejados estan los datos con la intencion de que no sea tan grande la diferencia

y=data3
Limite_superior=y.mean()+3*y.std()
Limite_inferior=y.mean()-3*y.std()
print('Limite superior permitido',Limite_superior)
print('Limite inferior permitido',Limite_inferior)
```

```python
#encontramos outliers del dataframe
outliers=data3[(y>Limite_superior)|(y<Limite_inferior)]
outliers
```

```python
#obtenemos dataos sin outliers
data4=data3[(y<=Limite_superior)&(y>=Limite_inferior)]
data4
```

```python
#corroboramos valores nulos
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos del data frame con mean
data5=data4.copy()
data5=data5.fillna(round(data4.mean(),1))
data5
```

```python
#corroboramos valores nulos
valores_nulos=data5.isnull().sum()
valores_nulos
```
