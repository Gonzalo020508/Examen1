```python
#hacemos data frame el dato importado
data=pd.read_csv('lastmile_delivery_operations_mit_lift_lab.csv')
data
```

```python
#verificamos informacion del data frame
data.info()
```

```python
#corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#realizamos una copia 
data1=data.copy()
```

**VALORES NULOS**

```python
#Al ser una gran cantidad de datos reemplazemos los valores nulos con "bfill"
data2=data1.fillna(method='bfill')
data2
```

```python
#verificamos nuestros valores nulos
valores_nulos1=data2.isnull().sum()
valores_nulos1
```

```python
#realizamos ahora un filtro por columnas de las que son de tipo cuantitativas 
datoscuantitativos=data2.iloc[ : , [5,6,12,21,22,25,28]]
datoscuantitativos
```

```python
#realizamos una copia del dataframe
data3=datoscuantitativos.copy()
```

```python
#corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#sustituir valores nulos con mediana
data3['serving_customer']=data3['serving_customer'].fillna(round(data2['serving_customer'].median(),1))
data3
```

```python
#corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#metodo de sustitucion de valores nulos por un valor concreto
data3['describe_the_accident']=data3['describe_the_accident'].fillna(0)
data3
```

```python
#corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#sustituir valores nulos por valores no nulos hacia adelate #forward fill
#filtro por columnas
data3['maximum_number_of_vehicles_in_the_traffic_jam']=data3['maximum_number_of_vehicles_in_the_traffic_jam'].fillna(method='ffill')
data3
```

```python
#corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

**PROCEDIMIENTO CUANTILES PARA SUSTITUIR OUTLIERS**

```python
#metodo para cuartiles encuentro cuartiles 0.25 y 0.75. El  método toma como referencia la diferencia entre el primer cuartil "Q1"  y el tercer cuartil "Q3"  se considera  un  valor  atípico  el  que  se  encuentra 1,5 veces esa distancia
#es un metodo "sencillo"
y=data1

percentile25=y.quantile(0.25)
percentile75=y.quantile(0.75)
iqr=percentile75-percentile25


Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos outliers del dataframe
outliers_iqr=data1[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#obtenemos datos limpios del data frama
data3_iqr=data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#realizamos ahora un filtro por columnas de las que son de tipo cuantitativas 
datoscuantitativos=data1.iloc[ : , [5,6,12,21,22,25,28]]
datoscuantitativos
```

```python
#corroboramos valores nulos
valores_nulos=datoscuantitativos.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atipicos del data frame con mean
data5_iqr=datoscuantitativos.copy()
data5_iqr=data5_iqr.fillna(round(datoscuantitativos.mean(),1))
data5_iqr
```

```python
#realizamos ahora un filtro por columnas de las que son de tipo cualitativas 
datoscualitativos=data1.iloc[ : , [0,1,2,3,4,7,8,9,10,11,13,14,15,16,17,18,19,20,23,24,26,27]]
datoscualitativos
```

```python
#concatenamos los datos cuantitativos y cualitativos
Datos_limpios=pd.concat([datoscualitativos,data5_iqr],axis=1)
Datos_limpios
```

```python
#corroboramos valores nulos
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```

```python
#Finalmente para no tener valores nulos en los object los sustituimos por string
#Reemplazar valor nulo de '_title' con texto 'Vacío'
Datos_limpios["_title"]=Datos_limpios["_title"].fillna("Vacío")
Datos_limpios
```

```python
#Reemplazar valor nulo de '_title' con texto 'Vacío'
Datos_limpios["plates"]=Datos_limpios["plates"].fillna("Vacío")
Datos_limpios
```

```python
#Reemplazar valor nulo de '_title' con texto 'Vacío'
Datos_limpios["company_if_visible"]=Datos_limpios["company_if_visible"].fillna("Sin empresa")
Datos_limpios
```

```python
#corroboramos valores nulos
valores_nulos=Datos_limpios.isnull().sum()
valores_nulos
```
